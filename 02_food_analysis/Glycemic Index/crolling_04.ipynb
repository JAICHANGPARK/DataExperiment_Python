```python
food_name = []
food_manufacturer = []
glycemic_index = []
standard_serve_size = []
carbohydrate_per_serve = []
glycemic_load = []
country = []
product_category = []
year_of_test = []
sem = []
time_period_of_test = []
num_of_subject = []
type_of_subject = []
source = []

for n in tqdm_notebook(df.index):
    url_sub = df['URL'][n]
    url = url_base + url_sub
    html = urlopen(url)
    soup = BeautifulSoup(html, "html.parser")
    tmp = soup.find('table', {'id' : 'table39'})
    a = tmp.find_all('tr')
    b = a[2].find_all('td')
    c = b[9:-2]

    fix_list = []
    for i in c:
        #print(i.get_text())
        fix_list.append(i.getText())
    fix_list = [x for x in fix_list if x != '\xa0']
    #print(fix_list[1::2])
    
    food_name.append(fix_list[1])
    food_manufacturer.append(fix_list[3])
    glycemic_index.append(fix_list[5])
    standard_serve_size.append(fix_list[7])
    carbohydrate_per_serve.append(fix_list[9])
    glycemic_load.append(fix_list[11])
    country.append(fix_list[13])
    product_category.append(fix_list[15])
    year_of_test.append(fix_list[17])
    sem.append(fix_list[19])
    time_period_of_test.append(fix_list[21])
    num_of_subject.append(fix_list[23])
    type_of_subject.append(fix_list[25])
    source.append(fix_list[27])
```

```python
food_manufacturer
```

```python
food_name = [fix_list[x] for x in oven]
```

```python
food_name
```

```python
data = {"Food Name":food_name, "Food Manufacturer":food_manufacturer, "GI":glycemic_index, 
       "Standard Serve Size": standard_serve_size, "Carbohydrate per Serve":carbohydrate_per_serve,
       "Glycemic Load":glycemic_load, "Country":country, "Product Category":product_category,
       "Year of Test": year_of_test, "SEM": sem, "Time Period of Test": time_period_of_test,
        "Number of Subjects in Test":num_of_subject, "Type of Subjects in Test":type_of_subject, "Source":source
       }
```

```python
df = pd.DataFrame(data)
```
