```python
total = pd.concat([tmp_beef,tmp_egg,tmp_fish])
```

```python
total.head()
```

```python
total = total.reset_index()
```

```python
del total['index']
```

```python
total.info()
```

```python
total.head()
```

```python
total = total.fillna(0.0)
```

```python
kcal = total['열량 (kcal)']
amount = total['1회제공량 (g)']
carbo = total['탄수화물 (g)']
protain = total['단백질 (g)']
fatt = total['지방 (g)']
```

```python
want_kcal = 75
```

```python
new_amount = (want_kcal * amount) / kcal 
new_carbo  = (want_kcal * carbo) / kcal 
new_protain  = (want_kcal * protain) / kcal 
new_fatt  = (want_kcal * fatt) / kcal 
```

```python
estimate_kcal = (new_carbo * 4) + (new_protain * 4) + (new_fatt * 9)
```

```python
total['열량 (kcal)'] = want_kcal
total['1회제공량 (g)'] = new_amount
total['탄수화물 (g)'] = new_carbo
total['단백질 (g)'] = new_protain
total['지방 (g)'] = new_fatt
```

```python
total.head(10)
```

```python
error = ( estimate_kcal - want_kcal ) 
total['Error'] = error
total.head()
```

```python
del total['당류 (g)']
del total['나트륨 (mg)']
del total['콜레스테롤 (mg)']
del total['포화지방산 (g)']
del total['트랜스지방산 (g)']
del total['년도']
```

```python
total.head()
```

```python
# 지방/ 단백질 
rate = 5 / 8
plus_rate = rate + rate * 0.1
print(plus_rate)
minus_rate = rate - (rate * 0.1)
print(minus_rate)
```

```python
new_rate =  new_fatt/ new_protain
```

```python
tmp = new_rate.between(minus_rate, plus_rate)
total['survived'] = tmp
```
