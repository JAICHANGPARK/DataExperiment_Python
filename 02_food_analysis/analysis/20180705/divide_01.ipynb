- 곡류군의 채워지지 않은 Nan 값을 0.0으로 초기화한다

```python
gokryu = gokryu.fillna(0.0)
```

열의 각 값을 가져와 시리즈로 저장한다

```python
kcal = gokryu['열량 (kcal)']
amount = gokryu['1회제공량 (g)']
carbo = gokryu['탄수화물 (g)']
protain = gokryu['단백질 (g)']
fatt = gokryu['지방 (g)']
```

100 kcal 에 대한 3대양소 비를 재 구성

```python
amount
```

```python
new_amount = (100 * amount) / kcal 
new_carbo  = (100 * carbo) / kcal 
new_protain  = (100 * protain) / kcal 
new_fatt  = (100 * fatt) / kcal 
```

```python
new_amount
```

```python
estimate_kcal = (new_carbo * 4) + (new_protain * 4) + (new_fatt * 9)
```

```python
gokryu['열량 (kcal)'] = 100
gokryu['1회제공량 (g)'] = new_amount
gokryu['탄수화물 (g)'] = new_carbo
gokryu['단백질 (g)'] = new_protain
gokryu['지방 (g)'] = new_fatt
```

```python
gokryu.head(30)
```

```python
estimate_kcal
```

```python
error = ( estimate_kcal - 100 ) 
```

```python
gokryu['Error'] = error
gokryu.head()
```

```python
del gokryu['당류 (g)']
del gokryu['나트륨 (mg)']
del gokryu['콜레스테롤 (mg)']
del gokryu['포화지방산 (g)']
del gokryu['트랜스지방산 (g)']
del gokryu['년도']
```

```python
gokryu.head()
```

```python
rate = 2 / 23
```

```python
plus_rate = rate + rate * 0.1
print(plus_rate)
minus_rate = rate - (rate * 0.1)
print(minus_rate)
```

```python
new_rate = new_protain / new_carbo
```

```python
type(new_rate)
```

```python
index = []
for i in range(len(new_rate)):
    if new_rate.between(minus_rate, plus_rate):
        index.append(i)
print(index)
```

```python
tmp = new_rate.between(minus_rate, plus_rate)
gokryu['survived'] = tmp
```
