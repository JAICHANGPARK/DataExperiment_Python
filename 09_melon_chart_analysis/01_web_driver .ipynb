```python
driver =webdriver.Chrome('./chromedriver.exe')
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)
```

```python
melon_html = driver.page_source
print(melon_html)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(melon_html, 'html.parser')
```

```python
soup
```

```python
songs = soup.select('#lst50')
```

```python
print(len(songs))
```

```python
print(songs[0])
```

```python
song = songs[0]
```

```python
print(song)
```

```python
title = song.select('a')
len(title)
```

```python
print(title)
```

```python
title = song.select('span > a')
len(title)
```
