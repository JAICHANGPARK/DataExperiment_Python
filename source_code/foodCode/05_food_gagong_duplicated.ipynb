```python
food3_duplicate = food03.drop_duplicates(['식품이름'], keep='last')
food3_duplicate
```

```python
food3_duplicate_count = food3_duplicate['식품군'].groupby(food3_duplicate['식품군']).describe()
food3_duplicate_count
```

```python
food3_duplicate_count['count'].sort_values().plot(kind='barh', grid= True, figsize = (10,15))
plt.xlabel('등장수')
plt.ylabel('KIND')
for i, v in enumerate(food3_duplicate_count['count'].sort_values()):
    plt.text(v, i, str(v), color='blue', fontweight='bold')
plt.show()
```
